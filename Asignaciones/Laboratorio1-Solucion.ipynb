```r
pool <- array(TRUE, dim = c(9,9,9))

#Revisamos cada entrada de s para actualizar las entradas de pool correspondientes
for(i in 1:9)
  for(j in 1:9){
    if(s[i,j] != 0){
      # Esto es para descartar todos los posibles valores en la casilla s[i,j]
      pool[i,j,1:9] <- FALSE
      # Esto es para descartar el valor s[i,j] de toda la columna j
      pool[1:9,j,s[i,j]] <- FALSE
      # Esto es para descartar el valor s[i,j] de toda la fila i
      pool[i,1:9,s[i,j]] <- FALSE
        
      # Ahora descartaremos el valor de s[i,j] del bloque de 3x3 donde esta
      index1 <- 3*(ceiling(i / 3) -1) + 1
      index2 <- 3*(ceiling(j / 3) -1) + 1
      for(a in 0:2)
        for(b in 0:2)
          pool[index1 + a, index2 + b, s[i,j]] <- FALSE
    }
  }
```

Comencemos por explicar que hace la función `trunc()`, esta función toma un argumento numerico y retorna el siguiente entero entre dicho numero y el 0, es como un `floor` para los positivos y un `ceiling` para los negativos. Como las entradas de la matriz están numeradas del 1 al 9, los bloques de 3x3 estarán divididos en los valores 1:3, 4:6 y 7:9. Tomando un $a$ cualquiera de 1:9, se cumplirá que:


 $$ 3k < a \leq 3(k+1),$$

como $a$ es un entero tenemos que:

  $$
  3k \leq a - 1 < 3(k+1) = k \leq (a-1)/3 < k+1,
  $$
 
Entonces, $k = trunc((a-1)/3) $.


De ello, `3*trunc((a-1)/3)` nos dará el valor del múltiplo de 3 que está inmediatamente antes que `a`. Los valores del bloque de `3` en el que esta contenido `a` son `3k+1`, `3k+2` y `3k+3` y ello es lo que se obtiene al sumar `1` a `boxa` y luego crea un vector desde ese valor hasta los dos siguientes.

Por tanto, `boxa` nos dá el bloque de `3` en el que está contenido `a` y `boxb` nos da el bloque de `3` en el que está contenido `b`  y al juntar ambos obtenemos el bloque de 3x3 en el que está contenida la entrada `(a,b)`.

```r
boxa <- 3 * trunc((a-1)/3) + 1
boxa <- boxa:(boxa + 2)
boxb <- 3 * trunc((b-1)/3) + 1
boxb <- boxb:(boxb + 2)
```

Debido a ello es que el bucle solo iterará sobre aquellos posibles valores para la entrada `s[a,b]` que aún no han sido descartados. Antes de continuar, cabe resaltar que en una suma de valores lógicos `(TRUE, FALSE)` a `TRUE` se le asocia el valor de `1` y `FALSE` el valor de `0`. La única instrucción dentro del bucle for verificará si dicho valor, `u`, aún es un candidato a la casilla `(a,b)` y modificará el valor de la entrada `pool[a,b,u]` a `FALSE` en caso ya no sea un candidato. 

Para ello se hace uso de `3` sumas de valores logicos, la primera verificará que el valor de `u` no se repita en ninguna otra entrada de la fila `a`, la segunda suma verifica que `u` no este presente en alguna otra entrada de la columna `b` y la última suma verifica el bloque de 3x3 en el que se encuentra la entrada `(a,b)`. ### Solución 8

Asumiremos que el palillo tiene longitud 1 (por la opción de unidades). Podemos generar los cuatro extremos de las partes y las longitudes de las partes de la siguiente manera.

```r
extremos = c(0, sort(runif(2)), 1)
longitudes = diff(extremos)
```

Los palillos se pueden ensamblar en un triángulo si la suma de las longitudes de las dos piezas más pequeñas es mayor que la longitud del más grande.

```r
longitudes = sort(longitudes)
sum(longitudes[1:2]) >= longitudes[3]
```

Todo lo que necesitamos hacer ahora es agrupar esto en una función de simulación.

```r
palillosim =function(n) {## seleccionamos los dos puntos de division 
    u1 = runif(n)
    u2 = runif(n)
    
## Obtenemos la matriz de puntos extremos del palillo
extremos = cbind(0, pmin(u1, u2), pmax(u1, u2), 1)
    
## Obtenemos las longitudes del palillo
l1 = extremos[,2] - extremos[,1]
l2 = extremos[,3] - extremos[,2]
l3 = extremos[,4] - extremos[,3]
    
## Prueba del triangulo
mean(l1 + l2 >= l3)
}

palillosim(10000000)
```

```r
# Problema de un juego de tenis. Solución dada por Victor Galvan Oyola.

p <- 0.3 # Inicializar la probabilidad de un punto
juegos <- 0.0 # Inicializar la probabilidad de ganar un juego
sets <- 0.0 # Inicializar la probabilidad de ganar un set
partidos <- 0.0 # Inicializar la probabilidad de ganar un partido
tiebreak <- 0.0 # Inicializar la probabilidad de ganar el tiebreak
tablaj <- matrix(0,nrow = 13,ncol=13) # Tabla para el DP para juegos
visj <- matrix(F,nrow = 13,ncol=13) # Tabla de estado visitado para el DP para juegos
tablas <- matrix(0,nrow = 7, ncol = 7) # Tabla para el DP para sets
viss <- matrix(F,nrow = 13,ncol=13) # Tabla de estado visitado para el DP para sets

# Se transforman los estados de i,j -> i+1,j+1 para que no haya contradiccion con los indices
# Los casos base (0,0) se transforman a (1,1) y se les asigna 1.0 como valor trivial de multiplicacion

dpj <- function(win, lose){
    if(win==1 && lose==1){ # Caso base
        return(1.0)
    }
    if(visj[win,lose]){ # Si ya se visito el estado
        return(tablaj[win,lose]) # Devolver el valor ya calculado
    }
    ans = 0.0 # Inicializar respuesta para este estado
    if(win == 1){ # Si no hay posibilidad de ganar mas puntos
        ans = ans + (1-p)*dpj(win,lose-1) # Se prueba solo con la probabilidad de perderlos
    } else if (lose == 1){ # Si no hay posibilidad de perder mas puntos
        ans = ans + p*dpj(win-1,lose) # Se prueba solo con la probabilidad de ganarlos
    } else{ # Si no son nulos
        ans = ans + p*dpj(win-1,lose)+(1-p)*dpj(win,lose-1) # Se prueba con la probabilidad de ganarlos o
                                                            # de perderlos hasta acabar todo
    }
    visj[win,lose] = T # Señalamos que el estado ya fue visitado
    tablaj[win,lose] = ans # Almacenamos la respuesta
    return(ans) # Devolvemos la respuesta
}

dps <- function(win, lose){
    if(win==1 && lose==1){  # Caso base
        return(1.0)
    }
    if(viss[win,lose]){ # Si ya se visito el estado
        return(tablas[win,lose]) # Devolver el valor ya calculado
    }
    ans = 0.0 # Inicializar respuesta para este estado
    if(win == 1){ # Si no hay posibilidad de ganar mas juegos
        ans = ans + (1-juegos)*dps(win,lose-1) # Se prueba solo con la probabilidad de perderlos
    } else if (lose == 1){ # Si no hay posibilidad de perder mas juegos
        ans = ans + juegos*dps(win-1,lose) # Se prueba solo con la probabilidad de ganarlos
    } else{ # Si no son nulos
        ans = ans + juegos*dps(win-1,lose)+(1-juegos)*dps(win,lose-1) # Se prueba con la probabilidad de ganarlos 
                                                                       # o de perderlos hasta acabar todo
    }
    viss[win,lose] = T # Señalamos que el estado ya fue visitado
    tablas[win,lose] = ans # Almacenamos la respuesta
    return(ans) # Devolvemos la respuesta
}

# Combinaciones para ganar en un juego: 
# 3, 0 y hacer el siguiente punto -> 4,1
# 3, 1 y hacer el siguiente punto -> 4,2
# 3, 2 y hacer el siguiente punto -> 4,3
# 3, 3 y hacer los dos siguientes puntos O hacer 1 punto, perder otro y verificar 
#la probabilidad de ganar en el caso 4,4
# En esta ultima tenemos la probabilidad de ganar en deuce con x puntos denotada por P(d(x)), entonces
#se nos da la igualdad:
# P(d(3)) = p*p + p*(1-p)*P(d(4)) + (1-p)*p*P(d(4))  -> El primer término se refiere a ganar 5-3, el segundo
#a la sucesion 3,3->4,3->4,4 y el ultimo a la sucesion 3,3->3,4->4,4 y el deuce en 4,4.
# Como el deuce no depende del valor actual de los puntos (3,3) tiene las mismas probabilidades
#que (4,4), entonces P(d(x)) = P(deuce):
# P(deuce) = p*p + 2*p*(1-p)*P(deuce) 
# P(deuce)*(1-2*p+2*p*p) = p*p
# P(deuce) =  p*p / (1-2*p+2*p*p)

for (i in c(1:3)){ # 3 primeras combinaciones
    juegos = juegos + p*dpj(4,i)
}
juegos = juegos + dpj(4,4)*p*p/(2*p*p-2*p+1) # 4ta combinacion

# Combinaciones para ganar en un set: 
# 5, 0 y hacer el siguiente punto -> 6,1
# 5, 1 y hacer el siguiente punto -> 6,2
# 5, 2 y hacer el siguiente punto -> 6,3
# 5, 3 y hacer el siguiente punto -> 6,4
# 5, 4 y hacer el siguiente punto -> 6,5
# 5, 5 y hacer los dos siguientes puntos O hacer un punto, perder otro y verificar 
#la probabilidad de ganar el tiebreak.
# El caso del TIEBREAK es similar a ganar un juego, pero en vez de 3 al inicio, se da con 6.
# Probabilidad de la 6ta combinacion se da como
# P = P(5,5)*juego*juego + P(5,5)*2*juego*(1-juego)*tiebreak

for (i in c(1:5)){ # Primeros 5 casos
    sets = sets + juegos*dps(6,i)
}
for (i in c(1:6)){ # Calculo del tiebreak para sus primeros 6 casos
    tiebreak = tiebreak + p*dpj(7,i)
}
tiebreak = tiebreak + dpj(7,7)*p*p/(2*p*p-2*p+1) # Tiebreak como la 4ta combinacion de juego
sets = sets + dps(6,6)*(juegos*juegos + 2*juegos*(1-juegos)*tiebreak) # 6ta combinacion para set
partidos = sets*sets + 2*sets*sets*(1-sets) # Secuencias 0,0->1,0->2,0 ; 0,0->1,0->1,1->2,1 ; 0,0->0,1->1,1->2,1

# Imprimimos los valores de cada respuesta
cat(sprintf(c("Juego=%.11f", "Set=%.11f","Partido=%.11f"), c(juegos,sets,partidos)), sep=" ") 
```
