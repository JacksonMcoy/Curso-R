## Eligiendo la base de datos IRIS

```r
# De las 150 linea elije 100 columnas aleatorias
itrain <- iris[sample(1:150,100),]
itrain $ setosa<- c( itrain $Species =='setosa')
itrain $ versicolor <- c( itrain $Species =='versicolor')
itrain $ virginica <- c( itrain $Species == 'virginica')
```

```r
itrain
itrain $ Species <- NULL
itrain
```

```r
inet <- neuralnet(setosa + versicolor + virginica ~ Sepal.Length + Sepal.Width + Petal.Length + Petal.Width, itrain, hidden=3, lifesign="full")
```

```r
plot(inet, rep="best")
```

```r
predict <- compute(inet, iris[1:4])
#predict
```

```r
which.max(predict$net.result[1,])
```

```r
result<-0
for (i in 1:150) { result[i] <- which.max(predict$net.result[i,]) }
```

```r
for (i in 1:150) { if (result[i]==1) {result[i] = "setosa"}}
for (i in 1:150) { if (result[i]==2) {result[i] = "versicolor"} }
for (i in 1:150) { if (result[i]==3) {result[i] = "virginica"} }
```

```r
comparison <- iris
comparison$Predicted <- result
comparison
```

# Aprendizaje NO Supervisado

```r
library("kohonen", lib.loc="~/R/x86_64-pc-linux-gnu-library/3.3")
library(kohonen)
i = iris
set.seed(7)
i$Species=NULL
i.sc = scale(i)
i.som = som(data = i.sc,grid = somgrid(5,5,topo = "hexagonal"))
```

* Ploteo - type = "codes"

Utilizando la funcion ploteo con el parametro type = "codes". No muestra la agrupacion de datos de acuerdo a sus cacteristicas, indicando cuales son estas que los indentifica a cada grupo. Adem√°s se puede observar la semejanza que existe en la interseccion de cada grupos, pareciendose a cada grupo de las intersecta.

```r
plot(i.som,main = "Datos de la tabla iris",type="codes")
```

* Ploteo - type = "changes"

Utilizando la funci√≥n ploteo con el parametro type = "changes". Muestra la distancia media al conjunto de elementos mas cercano, el cual cuanto mas peque√±o es esta media nos quiere decir que un elemento se encuentra cerca de otros elementos que poseen caracteristicas en com√∫n.

```r
plot(i.som,main = "Datos de la tabla iris",type="changes")
```

* Ploteo - type = "counts"

Utilizando la funcion ploteo con el paramentro type = "counts". Nos indica la cantidad de elementos que se encuentra en cada uno de los grupos. Si bien es cierto se podria preferir un distribucion uniforme para cada tama√±o que se asigne a la malla pero no es asi, dependiendo del tama√±o de la malla se visualizara la existencia de grupos vacios, para evitar esto es necesario disminuir la dimension de la malla. Por lo contrario si casi todos los grupos posee la maxima cantidad de elementos, para una mejor visualizacion es necesario aumentar la dimension de la malla.

```r
plot(i.som,main = "Datos de la tabla iris",type="counts")
```

* Ploteo - type = "dist.neighbours"

Utilizando la funcion ploteo con el paramentro type = "dist.neighbours".Nos ayuda a identificar una mayor agrupaci√≥n. El cual nos soluciona el problema que se tenia anteriormente. Y debido a la mayor distribuci√≥n se observa una disminuci√≥n en la maxima cantidad que puede haber en un grupo.

```r
plot(i.som,main = "Datos de la tabla iris",type="dist.neighbours")
```
